```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
face_classifier = cv2.CascadeClassifier('Haarcascades/haarcascade_frontalface_default.xml')
```

```python
img = cv2.imread('image.jpg')
fix_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(fix_img)
```

```python
face = face_classifier.detectMultiScale(img, 1.3, 5)

if face is ():
    print("No faces Found")
```

```python
def detect_face(fixed_image):
    face_rects = face_classifier.detectMultiScale(fixed_image)

    for (x, y, w, h) in face_rects:
        cv2.rectangle(fixed_image, (x,y), (x+w,y+h), (0,255,0),1)

    return fixed_image
```

```python
result = detect_face(fix_img)
plt.imshow(result)
```
